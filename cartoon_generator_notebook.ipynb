## Import Libraries

```
import torch
from diffusers import StableDiffusionImg2ImgPipeline
from PIL import Image
import gradio as gr
```

## Load Model

```
def load_model():
    """Load diffusion model"""
    model_id = "runwayml/stable-diffusion-v1-5"
    pipe = StableDiffusionImg2ImgPipeline.from_pretrained(model_id)
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")
    return pipe

pipe = load_model()
```

## Preprocess Image

```
def preprocess_image(image):
    """Resize & normalize image"""
    return image.convert("RGB").resize((512, 512))
```

## Cartoon Conversion

```
def convert_to_cartoon(image):
    """Convert image to cartoon"""
    image = preprocess_image(image)
    result = pipe(prompt="cartoon style, Pixar, anime", image=image, strength=0.75)
    return result.images[0]
```

## Save Output

```
def save_image(image, path="cartoon.png"):
    """Save image"""
    image.save(path)
    return path
```

## Gradio UI

```
def app(image, groq_key):
    """Gradio interface function"""
    output = convert_to_cartoon(image)
    save_image(output)
    return output

with gr.Blocks(theme=gr.themes.Soft(primary_hue="indigo")) as demo:
    gr.Markdown("# 🎨 Cartoon Generator")
    img = gr.Image(type="pil")
    out = gr.Image()
    key = gr.Textbox(label="Groq API Key", type="password")
    btn = gr.Button("Generate")
    btn.click(app, inputs=[img, key], outputs=out)

demo.launch()
```
